✅ 1. Import Libraries and Load Data

- Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

- Loading the Dataset

```python
df = pd.read_csv("../data/uci-energy-efficiency-dataset.csv")
```

- Renaming Columns for Readability

```python
df.columns = ['Relative_Compactness', 'Surface_Area', 'Wall_Area', 'Roof_Area',
              'Overall_Height', 'Orientation', 'Glazing_Area', 'Glazing_Area_Distribution',
              'Heating_Load', 'Cooling_Load']
```

- `df.info()`

```python
df.info()
```

Explanation:
  - The info() function is used to get a summary of the DataFrame.
  - It returns the following details:
    1. Number of entries (rows): How many rows are there in the dataset.
    2. Column names: A list of all column names in the dataset.
    3. Non-null values: The number of non-null (non-missing) values in each column.
    4. Data types: The data type (e.g., int, float, object) for each column.

- `df.describe()`

```python
df.describe()
```

Explanation:
  - The describe() function provides summary statistics for each numerical column in the    DataFrame. Specifically, it gives:
    1. Count: Number of non-null values.
    2. Mean: The average value of the column.
    3. Standard Deviation (std): Measure of the spread of data (how much data deviates from the mean).
    4. Min: The smallest value.
    5. 25% (Q1): The 25th percentile (lower quartile).
    6. 50% (Q2/Median): The 50th percentile (the middle value when data is sorted).
    7. 75% (Q3): The 75th percentile (upper quartile).
    8. Max: The largest value.

- `df.isnull().sum()`

```python
df.isnull().sum()
```

Explanation:
- df.isnull(): This checks for missing values (NaNs) in the DataFrame. It returns a new DataFrame of the same shape, but with True for missing values and False for non-missing values.
- sum(): After calling isnull(), applying sum() will count the number of missing values in each column.

✅ 2. Visualizing the Distribution of Heating Load with Histogram and KDE

```python
# Set the figure size for the plot (8 inches wide, 6 inches tall)
plt.figure(figsize=(8, 6))

# Create a histogram with a Kernel Density Estimate (KDE) overlay for the 'Heating_Load' column
# 'bins=30' specifies the number of bins for the histogram
sns.histplot(df['Heating_Load'], kde=True, bins=30)

# Set the title of the plot with a font size of 14
plt.title("Heating Load Distribution", fontsize=14)

# Set the label for the x-axis (Heating Load) with a font size of 12
plt.xlabel("Heating Load", fontsize=12)

# Set the label for the y-axis (Count of occurrences) with a font size of 12
plt.ylabel("Count", fontsize=12)

# Automatically adjust the layout to ensure everything fits within the figure area
plt.tight_layout()

# Display the plot
plt.show()
```

✅ 3. Visualizing the Distribution of Cooling Load with Histogram and KDE

```python
# Set the figure size for the plot (8 inches wide, 6 inches tall)
plt.figure(figsize=(8, 6))

# Create a histogram with a Kernel Density Estimate (KDE) overlay for the 'Cooling_Load' column
# 'bins=30' specifies the number of bins for the histogram
sns.histplot(df['Cooling_Load'], kde=True, bins=30)

# Set the title of the plot with a font size of 14
plt.title("Cooling Load Distribution", fontsize=14)

# Set the label for the x-axis (Cooling Load) with a font size of 12
plt.xlabel("Cooling Load", fontsize=12)

# Set the label for the y-axis (Count of occurrences) with a font size of 12
plt.ylabel("Count", fontsize=12)

# Automatically adjust the layout to ensure everything fits within the figure area
plt.tight_layout()

# Display the plot
plt.show()
```

✅ 4. Feature Correlation

```python
# Set the figure size for the heatmap (12 inches wide, 8 inches tall)
plt.figure(figsize=(12, 8))

# Create a heatmap to visualize the correlation matrix of the DataFrame
# df.corr() computes the correlation between numerical features
# annot=True displays the correlation values inside the heatmap cells
# cmap="coolwarm" specifies the color map to be used (from cool to warm colors)
# fmt=".2f" formats the correlation values to 2 decimal places
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")

# Set the title of the plot with a font size of 14
plt.title("Feature Correlation Matrix", fontsize=14)

# Automatically adjust the layout to ensure everything fits within the figure area
plt.tight_layout()

# Display the plot
plt.show()
```

✅ 5. Outlier Detection

```python
# Select all numeric columns in the DataFrame using pandas' select_dtypes method
# np.number selects columns with numeric data types (int, float, etc.)
numeric_cols = df.select_dtypes(include=np.number).columns

# Loop through each numeric column in the DataFrame
for col in numeric_cols:
    # Create a new figure with a specified size (6 inches wide, 3 inches tall) for each boxplot
    plt.figure(figsize=(6, 3))
    
    # Create a boxplot for the current numeric column (col)
    sns.boxplot(x=df[col])
    
    # Set the title for the boxplot dynamically using the column name
    plt.title(f"{col} Boxplot")
    
    # Automatically adjust the layout to ensure everything fits properly
    plt.tight_layout()
    
    # Display the plot
    plt.show()
```

✅ 6. Categorical Analysis

```python
# Create a count plot to visualize the distribution of the 'Orientation' column
# 'x='Orientation'' sets the 'Orientation' column on the x-axis
# 'data=df' specifies that the data comes from the DataFrame 'df'
# 'hue='Orientation'' ensures that different 'Orientation' categories are shown in different colors in the plot
sns.countplot(x='Orientation', data=df, hue='Orientation')

# Set the title of the plot with a font size of 14
plt.title("Orientation Distribution", fontsize=14)

# Set the label for the x-axis (Orientation categories) with a font size of 12
plt.xlabel("Orientation", fontsize=12)

# Set the label for the y-axis (Count of occurrences) with a font size of 12
plt.ylabel("Count", fontsize=12)

# Automatically adjust the layout to ensure everything fits within the figure area
plt.tight_layout()

# Display the plot
plt.show()
```
